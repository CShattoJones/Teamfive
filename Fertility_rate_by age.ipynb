```python
# Modules

# Load module to create Panda DataFrames in Jupyter Notebooks
import pandas as pd
# Load module to create and format various plots
import matplotlib.pyplot as plt
# Load module to handle high-level math functions and basic algebra
import numpy as np
# Load module to allow communication with HTTPS for API requests
import requests
# Load module that deserializes java script open notation into Jupyter Notebooks
import json
# Load module that returns the number of seconds passed since epoch 
import time
# Load module that contains probability distributions and statistical functions
from scipy.stats import linregress
```

```python
# Total Fertility by Country
# Define variable name that holds csv file location of the fertility rates
tot_fertility_df = "testupdate.csv"
# Define variable name the holds the read information from the csv file
tot_fertility_df = pd.read_csv(tot_fertility_df)
# len(fertility_clean_df) 1043 
# fertility_clean_df['Country'].nunique() 233
# fertility_clean_df.head(25)
# fertility_clean_df.to_csv("test.csv")
tot_fertility_df.head(25)
```

```python
#Total Fertility by Agees
age_fert_df = tot_fertility_df.loc[((tot_fertility_df["Reference Year"]==2005) | (tot_fertility_df["Reference Year"]==2006)
                                           |(tot_fertility_df["Reference Year"]==2007)) |(tot_fertility_df["Reference Year"]==2008)
                                            |(tot_fertility_df["Reference Year"]==2009)]


age_fert_df = age_fert_df.groupby('Reference Year').agg({'15-19': ['mean'], '20-24': ['mean'], '25-29': ['mean'],
                                                  '30-34': ['mean'], '35-39': ['mean'], '40-44': ['mean'],
                                                  '45-49': ['mean']})
age_fert_df.head(25)
```

```python
age_fert_df.plot.bar(stacked=True, figsize = (10,10))
plt.xlabel("Year")
plt.ylabel("Total fertility (live births per woman)")
plt.title("For All Countries - Total Fertility by Ages")
plt.grid(True, linestyle='--')
plt.savefig('Age_fert_rate_chart', transparent=True, bbox_inches='tight')
plt.show()                     
```

```python
#Total Fertility Rate
one_fert_df = tot_fertility_df.loc[((tot_fertility_df["Reference Year"]==2005) | (tot_fertility_df["Reference Year"]==2006)
                                           |(tot_fertility_df["Reference Year"]==2007)) |(tot_fertility_df["Reference Year"]==2008)
                                            |(tot_fertility_df["Reference Year"]==2009)]


one_fert_df = one_fert_df.groupby('Reference Year').agg({'Total Fertility': ['mean']})
one_fert_df
```

```python
one_fert_df.plot.bar(stacked=True, figsize = (10,10))
plt.xlabel("Year")
plt.ylabel("Total fertility (live births per woman)")
plt.title("For All Countries - Total Fertility")
plt.grid(True, linestyle='--')
plt.savefig('One_fert_rate_chart', transparent=True, bbox_inches='tight')
plt.show()    
```

```python
# Fertility CSV:

# Define variable name that holds csv file location of the fertility rates
fertility_clean_df = "clean_data/fertility_clean.csv"
# Define variable name the holds the read information from the csv file
fertility_clean_df = pd.read_csv(fertility_clean_df)
# len(fertility_clean_df) 1043 
# fertility_clean_df['Country'].nunique() 233
# fertility_clean_df.head(25)
# fertility_clean_df.to_csv("test.csv")
fertility_clean_df.dtypes
```

```python
# Age specific birth rate per 1000 women
#fertility_Refyr_df = fertility_clean_df.loc[fertility_clean_df["Reference Year"]==2006]

fertility_Refyr_df = fertility_clean_df.loc[((fertility_clean_df["Reference Year"]==2005) | (fertility_clean_df["Reference Year"]==2006)
                                           |(fertility_clean_df["Reference Year"]==2007)) |(fertility_clean_df["Reference Year"]==2008) 
                                            |(fertility_clean_df["Reference Year"]==2009) ]

# Use pandas column rename function (.columns) to rename the columns in the dataframe
fertility_Refyr_df.columns=  ['Country', 'ISO Code','Reference Year','Total Fertility',
                              '15-19', '20-24', '25-29', '30-34', '35-39','40-44','45-49', 'Mean Age at Childbearing']
fertility_Refyr_df = fertility_Refyr_df.dropna(how='any')
fertility_Refyr_df.reset_index(drop=True, inplace=True)
fertility_Refyr_df= fertility_Refyr_df.loc[fertility_Refyr_df['45-49'] != '..']
fertility_Refyr_df['45-49'] = fertility_Refyr_df['45-49'].astype(float)
age_mean_df = fertility_Refyr_df.groupby('Reference Year').agg({'15-19': ['mean'], '20-24': ['mean'], '25-29': ['mean'],
                                                  '30-34': ['mean'], '35-39': ['mean'], '40-44': ['mean'],
                                                  '45-49': ['mean']})
age_mean_df
```

```python
age_mean_df.plot.barh(stacked=True, figsize = (10,10))
plt.xlabel("Year")
plt.ylabel("Number of Live Births by Age group per 1000 women")
plt.title("For All Countries - Age Specfic Birth Rate")
plt.grid(True, linestyle='--')
plt.savefig('Age_spec_rate_chart', transparent=True, bbox_inches='tight')
plt.show()
```
